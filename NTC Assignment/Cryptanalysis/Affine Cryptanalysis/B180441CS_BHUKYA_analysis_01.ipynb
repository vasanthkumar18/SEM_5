```python
def inverse_mod(l): # checking if there's inverse for l in Z26
    for i in range(1,26):
        if (l*i)%26==1:
            return i
        #we pass the inputs only which have the inverse and finally i is the inverse which we are looking for
```

```python
def get_key(plain,cipher):
    
    for i in range(26):
        k1=(i*((ord(plain[0])-ord(plain[1]))))%26
        k2=(ord(cipher[0])-ord(cipher[1]))
        
        if k1==k2:
            k=i
            
            x=ord(cipher[0])-((ord(plain[0])*k)%26)
            
            for j in range(26):
                
                if j%26==x%26:
                    break
    return k,j
# getting the key from plain and cipher
```

```python
a=3
b=5
def enc_affine(plain):
    result=""
    for c in plain:
        if c.isupper():         # incase the character is UPPERCASE
            d=(ord(c)-65)*a+b
            result+=chr((d%26)+65)
        elif c.islower():       # incase the character is UPPERCASE
            d=(ord(c)-97)*a+b
            result+=chr((d%26)+65)
        else:                   # space (considering proper message)
            result=result+''
    return result  
#result has the encrypted text
```

```python
a=3
b=5
def dec_affine(cipher):
    result=""
    for c in cipher:
        if c.isupper():
            d=((ord(c)-65)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
        elif c.islower():
            d=((ord(c)-97)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
    return result   
 #result has the decrypted text
```

```python
def dec_affinekey(cipher,a,b):
    result=""
    for c in cipher:
        if c.isupper():
            d=((ord(c)-65)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
        elif c.islower():
            d=((ord(c)-97)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
    return result 
 #result has the decrypted text
```
