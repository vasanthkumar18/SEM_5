```python
def inverse_mod(l): # checking if there's inverse for l in Z26
    for i in range(1,26):
        if (l*i)%26==1:
            return i
    #we pass the inputs only which have the inverse and finally i is the inverse which we are looking for
```

```python
def get_key(plain,cipher):
    d=ord(cipher[0])-ord(plain[0])
    return d
# getting the key from plain and cipher
```

```python
b=14
def enc_shift(plain,a):
    result=""
    for c in plain:
        if c.isupper():         # incase the character is UPPERCASE
            d=(ord(c)-65)*a+b
            result+=chr((d%26)+65)
        elif c.islower():       # incase the character is UPPERCASE
            d=(ord(c)-97)*a+b
            result+=chr((d%26)+65)
    return result
#result has the encrypted text
```

```python
b=14
def dec_shift(cipher,a):
    result=""
    for c in cipher:
        if c.isupper():
            d=((ord(c)-65)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
        elif c.islower():
            d=((ord(c)-97)-b)*inverse_mod(a)
            result+=chr((d%26)+65)
    return result
 #result has the decrypted text 
```

```python
def dec_shiftkey(cipher,key):
    plainmessage=""
    a=1
    for c in cipher:
        if c.isupper():
            d=((ord(c)-65)-key)*inverse_mod(a)
            plainmessage+=chr((d%26)+65)
        elif c.islower():
            d=((ord(c)-97)-key)*inverse_mod(a)
            plainmessage+=chr((d%26)+65)
    return plainmessage
 #result has the decrypted text 
```
